```python
import numpy as np
from fcmeans import FCM
import matplotlib.pyplot as plt
```

```python

#data values
sample=4000
data_points=np.concatenate((np.random.normal((-2,-2),size=(sample,2)),
np.random.normal((-2,-2),size=(sample,2))))
```

```python
len(data_points)
```

```python
#instance fcm
fc = FCM(n_clusters=2)
fc.fit(data_points)
```

```python
centroids = fc.centers
```

```python
fc.predict(data_points)
```

```python
#plot the graph
fig,axes=plt.subplots(1,2,figsize=(11,5))
axes[0].scatter(data_points[:,0],data_points[:,1],alpha=0.1)
axes[1].scatter(data_points[:,0],data_points[:,1],c=fc.predict(data_points),alpha=0.1)
axes[1].scatter(centroids[:,0],centroids[:,1],marker='*',c='w',s=400)
plt.show()
```
