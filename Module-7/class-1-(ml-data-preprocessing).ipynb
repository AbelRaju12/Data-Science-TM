```python
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(np.arange(16, 32).reshape(4,4), columns=['a','b','c','d'])
df1
```

```python
df2 = pd.DataFrame(np.arange(9).reshape(3,3), columns=['a','b','c'])
df2
```

```python
final_df = pd.concat([df1, df2])
```

```python
final_df
```

```python
si = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
imputed = si.fit_transform(final_df)
```

```python
pd.DataFrame(imputed, columns=['a','b','c','d'])
```
