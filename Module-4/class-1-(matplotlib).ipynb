```python
import pandas as pd
df = pd.read_csv('Expanded_data_with_more_features.csv')
df.head()
```

```python
df.groupby('EthnicGroup')['MathScore'].mean()
```

```python
for grp, data in df.groupby('EthnicGroup'):
    plt.bar(grp,data['MathScore'].mean())
plt.title('Math Scores')
plt.xlabel('Ethnic Group')
plt.ylabel('Math score')
```

```python
data = df[10:15]
width = 0.23
plt.figure(figsize=(10,10))
plt.bar(data.index-width,data['MathScore'],width=width)
plt.bar(data.index,data['ReadingScore'],width=width)
plt.bar(data.index+width,data['WritingScore'],width=width)
```
