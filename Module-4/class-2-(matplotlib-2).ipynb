```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
df = pd.read_csv('Expanded_data_with_more_features.csv')
df.head()
```

```python
#pie - chart

data = df['EthnicGroup'].value_counts()
```

```python
type(data)
```

```python
data.index
```

```python
plt.pie(data, labels = data.index)
plt.show()
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
plt.pie(data, labels = data.index,autopct='%.2f%%', explode = [0, 0.3, 0.7, 0, 0.1])
plt.show()
```
