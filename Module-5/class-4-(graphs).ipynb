```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cereal = pd.read_csv('cereal.csv')
cereal.head()
```

```python
plt.figure(figsize = (20, 5))
# cereal.columns[3:11]
index = 1
for i in cereal.columns[3:11]:
    plt.subplot(1, 8, index)
    sns.histplot(cereal[i])
    index += 1
```

```python
#1 compare manf. based on customer ratings
#2 which manufactureres are the largest producers of cereals
```

```python
cereal.groupby('mfr')['rating'].mean().plot.bar()
plt.xlabel('manufacturers')
plt.ylabel('ratings')
plt.title('Manufacturers with highest ratings')
plt.show()
```

```python
data = cereal['mfr'].value_counts()
data.index
data.values
```

```python
plt.pie(data.values, labels = data.index)
plt.show()
```
