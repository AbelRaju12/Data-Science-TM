```python
air = pd.read_csv('Airbnb_Open_Data.csv')
air.head()
```

```python
air['neighbourhood'].value_counts()
```

```python
harlem = air.apply(lambda x: True if x['neighbourhood'] == 'Harlem' else False, axis = 1)
```

```python
har = len(harlem[harlem == True].index)
```

```python
p_har = round(har/10234, 3)
p_har
```

```python
#conditional probability
data = pd.DataFrame({'count':air.groupby(['neighbourhood group', 'neighbourhood']).size()}).reset_index()
data
```

```python
data[data['neighbourhood group'] == 'Bronx']
```

```python
ss = data[data['neighbourhood group'] == 'Bronx']['count'].sum()
ss
```

```python
p_apt  = round(1000 / ss, 3)
p_apt
```
