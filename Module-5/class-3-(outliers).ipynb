```python
import pandas as pd
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv('glass.csv')
df.head()
```

```python
df.shape
```

```python
#z score
z_scores = stats.zscore(df)
```

```python
threshhold = 2
np.where(z_scores > threshhold)
```

```python
df[(z_scores < threshhold).all(axis = 1)]
```
